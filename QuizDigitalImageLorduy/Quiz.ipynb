```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading image files
img1 = cv2.imread('Imagenesquiz/imagen1.jpg')
img2 = cv2.imread('Imagenesquiz/imagen2.jpg')
img3 = cv2.imread('Imagenesquiz/imagen3.jpg')
img4 = cv2.imread('Imagenesquiz/imagen4.jpg')
dim = (680,480)
img1r = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
img2r = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)
img3r = cv2.resize(img3, dim, interpolation = cv2.INTER_AREA)
img4r = cv2.resize(img4, dim, interpolation = cv2.INTER_AREA)

# Applying NumPy addition on images

fimg = img1r + img2r + img3r + img4r
 
# Saving the output image
plt.imshow(fimg,cmap='gray')
plt.show()
```
